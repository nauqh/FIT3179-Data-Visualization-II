```python
df = pd.read_csv("https://raw.githubusercontent.com/albertyw/avenews/master/old/data/average-latitude-longitude-countries.csv")
df = df.rename(columns={'ISO 3166 Country Code': 'Code'})
df
```

```python
netflix = pd.read_csv("https://raw.githubusercontent.com/nauqh/FIT3179-Data-Visualization-II/main/data/titles.csv")
netflix.dropna(subset=['production_countries'])
netflix['production_countries'] = netflix['production_countries'].apply(literal_eval)
netflix = netflix[netflix['production_countries'].apply(lambda x: len(x) > 0)]
netflix['production_countries']=netflix['production_countries'].map(lambda x: x[0])
netflix = netflix.rename(columns={'production_countries': 'Code'})
netflix
```

```python
merged_df = pd.merge(netflix, df, on='Code')
```

```python
merged_df.to_csv("netflix2.csv", index=False)
```
